**Problem** - Write down a pseudo-code to parse the coordinate file and output the torsional (dihedral) angles

*Torsional angle* -   


**Solution** -  
Let's assume, the coordinates of the points(elements) A, B, C and D are given as-

$\overrightarrow{\text{A}}$ - `(ax,ay,az)`  
$\overrightarrow{\text{B}}$ - `(bx,by,bz)`  
$\overrightarrow{\text{C}}$ - `(cx,cy,cz)`  
$\overrightarrow{\text{D}}$ - `(dx,dy,dz)`  

* The dihedral angle is the angle between two planes consisting the points A, B, C and B, C, D.  

* The angle between the two planes can be calculated by calculating the angle between the respective normal of the two planes. *[Dot Product of the Normals]*  

* Normal of the plane can be found by cross multiplication of two lines lying on the plane.  

The normal of the plane consisting the points A, B and C can be calculated as-  
$\overrightarrow{\text{AB}} \times \overrightarrow{\text{CB}}$  

The normal of the plane consisting the points B, C and D can be calculated as-  
$\overrightarrow{\text{BC}} \times \overrightarrow{\text{DC}}$

*Dot Product*-  
  

*Cross Product*-  


```python
# CODE
import numpy as np
import math

# coordinates
a_cap = np.array([1,0,0]) # a point on x-axis
b_cap = np.array([0,0,0]) # origin
c_cap = np.array([0,1,0]) # a point on y-axis
d_cap = np.array([0,1,1]) # a point in y-z plane 
# expected torsional angle- 90 degrees

# finding two line vectors
ab_cap = b_cap - a_cap # line AB
cb_cap = b_cap - c_cap # line CB
bc_cap = -1 * cb_cap 
dc_cap = c_cap - d_cap # line DC

# normal of planes
abc_plane = np.cross(ab_cap,cb_cap)
bcd_plane = np.cross(bc_cap,dc_cap)

# torsionl angle calculation

def tau_angle(array1, array2):
    abs1, abs2 = 0,0
    for i in range(3):
        abs1 += array1[i] ** 2
        abs2 += array2[i] ** 2
    abs1 = math.sqrt(abs1) # absolute value of vector
    abs2 = math.sqrt(abs2)

    tau = math.acos(np.dot(array1,array2)/(abs1*abs2)) # look dot product formula
    return math.degrees(tau)
```

```python
print(tau_angle(abc_plane,bcd_plane))
```
